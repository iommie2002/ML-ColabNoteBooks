```python
import pandas as pd

df=pd.read_csv('aapl_raw_data.csv')
df.head()
```

```python
df.tail()
```

```python
from matplotlib import pyplot as plt
plt.plot(df.open, df.high)
plt.show()
```

```python
plt.bar(df.close ,df.change_percent)


plt.title("change percent at closing volume")
plt.ylabel("change percent")
plt.xlabel("close")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['close'], df['change_percent'], alpha=0.5)
plt.title("Change Percent vs. Closing Price")
plt.xlabel("Closing Price")
plt.ylabel("Change Percent")
plt.grid(True)
plt.show()
```
