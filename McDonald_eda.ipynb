```python
#converting servesize and calories to numerical data type
df['servesize'] = df['servesize'].astype('float64')
df['calories'] = df['calories'].astype('float64')

df.info()
```

```python
unique_values = df['protien'].value_counts()
print(unique_values)
```

```python
x = unique_values.index  # Unique protein values
y = unique_values.values  # Counts of each unique protein value

plt.figure(figsize=(10, 6))
plt.scatter(x, y, marker='o', alpha=0.5)
plt.xlabel('Protein Value')
plt.ylabel('Count')
plt.title('Scatter Plot of Unique Protein Values')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

plt.show()
```

```python
unique_values=df['cholestrol'].value_counts()
print(unique_values)
```

```python
cholesterol_data = df['cholestrol']

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(cholesterol_data, vert=True)  # vert=False for horizontal box plot
plt.xlabel('Cholesterol Level')
plt.title('Box Plot of Cholesterol')

plt.show()
```

```python
sodium_data = df['sodium']

# Create a line graph
plt.figure(figsize=(8, 6))
plt.plot(sodium_data, marker='o', linestyle='-')
plt.xlabel('Data Point Index')
plt.ylabel('Sodium Level')
plt.title('Line Graph of Sodium')

plt.show()
```

```python

sugar_data = df['sugar']
carbs_data = df['carbs']

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(sugar_data, carbs_data, alpha=0.5)  # Alpha sets the transparency of points
plt.xlabel('Sugar Content')
plt.ylabel('Carbs Content')
plt.title('Scatter Plot of Sugar vs Carbs')

plt.show()
```
