```python
#list and tupple are  ordered
#set is not ordered and hence  element cannot be accessed  hence we need to use for loop

#this is set1


seta={"apple","mango","watermelon","banana","pomegranate"}

#this is set2
setb={"apple","mango","watermelon","banana","pomegranate","swift","thar","endeavour"}
print(seta)
print(setb)

#intersection
print("intersection")
set1=seta.intersection(setb)
print(set1)
print(seta.intersection_update(setb))

#difference
print("set a.b difference")
set2=seta.difference(setb)
print(set2)

print("set b.a difference")
set3=setb.difference(seta)
set4=setb.difference_update(seta)
print(set3)
print(set4)

#disjoint
print("is set a disjoint of b ?")
print(seta.isdisjoint(setb))

print("is set b disjoint of a ?")
print(setb.isdisjoint(seta))

#subset
print("is set a subset of b ?")
print(seta.issubset(setb))

print("is set b subset of a ?")
print(setb.issubset(seta))

#superset
print("is set a superset of b ?")
print(seta.issuperset(setb))

print("is set b superset of a ?")
print(setb.issuperset(seta))


```
