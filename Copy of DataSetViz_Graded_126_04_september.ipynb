```python
import pandas as pd
df=pd.read_csv('Attrition1.csv')
df.head()
```

```python
#Question 1 :   Plot Pie Chart for Business Travel



from matplotlib import pyplot as plt
count = df['BusinessTravel'].value_counts()
plt.figure(figsize=(8, 8))

plt.pie(count, labels=count.index, autopct='%1.1f%%', startangle=140)
plt.title("Pie chart for business travel")
plt.grid(True)
plt.show()
```

```python
#Question 2 :

#Plot the above pie chart in yello,blue and green for the given categories

count = df['BusinessTravel'].value_counts()

# Define colors for each category
colors = ['yellow', 'blue', 'green']

plt.figure(figsize=(8, 8))

plt.pie(count, labels=count.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title("Pie chart for business travel")
plt.grid(True)
plt.show()
```

```python
#Question 3 :

#Plot the Bar Graph for Gender vs. Business Travel


import matplotlib.pyplot as plt
import pandas as pd



# Count the occurrences of each combination of Gender and BusinessTravel
count = df.groupby(['Gender', 'BusinessTravel']).size().unstack(fill_value=0)

# Plot the bar graph
count.plot(kind='bar', stacked=False, figsize=(10, 6))

# Set labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Graph of Gender vs. Business Travel')

# Show the legend
plt.legend(title='Business Travel')

# Show the plot
plt.show()
```

```python
#Question 4 :
#Plot any graph in stacked manner


# Count the occurrences of each combination of Gender and BusinessTravel
count = df.groupby(['Gender', 'BusinessTravel']).size().unstack(fill_value=0)

# Plot the bar graph
count.plot(kind='bar', stacked=True, figsize=(10, 6))

# Set labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Graph of Gender vs. Business Travel')

# Show the legend
plt.legend(title='Business Travel')

# Show the plot
plt.show()
```

```python
#Question 5:
#Plot pie chart graph for job roles

count = df['JobRole'].value_counts()
plt.figure(figsize=(8, 8))

plt.pie(count, labels=count.index, autopct='%1.1f%%', startangle=140)
plt.title("Job Role")
plt.grid(True)
plt.show()
```

```python
#Question 6:

#plot the box plot for years at the  company

plt.figure(figsize=(8, 6))
plt.boxplot(df['YearsAtCompany'], vert=True)  # vert=False for a horizontal box plot
plt.title('Box Plot of Years at Company')
plt.xlabel('Years')
plt.grid(True)

# Show the plot
plt.show()
```
