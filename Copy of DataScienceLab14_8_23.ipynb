```python
arr1=[1,2,3,4]
arr2=[5,6,7,8]
arr3=np.concatenate((arr1,arr2))
print(arr3)


arr4=[[1,2],[3,4],[5,6],[7,8]]
print(arr4)
arr5=[[11,12] ,[13,14] ,[15,16] ,[17,18]]
print(arr5)

arr6=np.concatenate((arr4,arr5),axis=1)
print(arr6)
```

```python
arr1=np.zeros((3,4))
print("zero matrix")

print(arr1)

arr2=np.ones((3,3))
print("ones matrix")
print(arr2)
arr3=np.identity((3))
print("identity matrix")
print(arr3)
```

```python
a =np.array([1,2,2,6,14,8,9,2,3])
print(a)

print(a[a>5])
print(a[a%2==0])


print(a[(a>5) & (a%2==0)])
print(np.where(a[(a>5) & (a%2==0)]))
```
