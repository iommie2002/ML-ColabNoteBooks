# **Removing "NEW" and 'Nove' from Rate Column**

```python
def handlerate(value):
    if(value=='NEW' or value=='Nove'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rating'] = df['rating'].apply(handlerate)
df['rating'].head()
```

```python
def handlecomma(value):
  value=str(value)
  if ',' in value:
    value=value.replace(',','')
    return float(value)
  else:
    return float(value)

df['cost_for_two']=df['cost_for_two'].apply(handlecomma)
df['cost_for_two'].unique()
```

```python
df['city'].value_counts()
```

```python
city=df['city'].value_counts(ascending=False)

city_lessthan300=city[city<300]

def handle_city(value):
  if(value in city_lessthan300):
    return 'others'
  else:
    return value

df['city']=df['city'].apply(handle_city)
df['city'].value_counts()
```
