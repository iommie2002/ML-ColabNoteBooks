### **5.Defining Independant Variable**

```python
# Putting feature variable to X
X = df['TV']

# Print the first 5 rows
X.head()
```

### **6.Defining Dependant Variable**

```python
# Putting response variable to y
y = df['Sales']

# Print the first 5 rows
y.head()
```

### **7.Splitting the Dataset**

```python
#random_state is the seed used by the random number generator, it can be any integer.

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7 , random_state=0000)
```

```python
print(type(X_train))
print(type(X_test))
print(type(y_train))
print(type(y_test))
```

```python
train_test_split
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
import numpy as np
#Simply put, numpy.newaxis is used to increase the dimension of the existing array by one more dimension,
X_train = X_train[:, np.newaxis]
X_test = X_test[:, np.newaxis]
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### **8.Linear Regression Model**

```python
# import LinearRegression from sklearn
from sklearn.linear_model import LinearRegression

# Representing LinearRegression as lr(Creating LinearRegression Object)
lr = LinearRegression()

# Fit the model using lr.fit()
lr.fit(X_train, y_train)
```

### **9.Evaluation Metrics**

```python
# Importing necessary libraries
from sklearn.metrics import mean_squared_error, r2_score

# Predict on the test data
y_pred = lr.predict(X_test)

# Print evaluation metrics
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R-squared:", r2_score(y_test, y_pred))
```

### **11.Actual vs Predicted Value**

### **12.Conclusion**


In conclusion, linear regression is a fundamental statistical technique used for modeling the relationship between a dependent variable and one or more independent variables. It aims to find the best-fitting line that represents the linear relationship between variables, making it suitable for predicting continuous outcomes. Although linear regression is simple to implement and interpret, it assumes a linear relationship between variables and may not perform well when the relationship is non-linear or when outliers are present. Nonetheless, it serves as a cornerstone in statistical modeling and is extensively employed in fields like economics, engineering, and social sciences for forecasting and trend analysis.
